```python
#Load Libraries
import glob
import numpy
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Load Dataset
dataset = sorted(glob.glob('data*.csv'))
```

```python
#Defining Function visualize
def visualize(dataset):
    
    #Read the dataset
    data = pd.read_csv(dataset)

    #Defining variable to hold data    
    average_data = []
    max_data = []
    min_data = []
    
    #For Loop to calculate the mean, max, and min for each column and add to the new variable 
    for stats in data:
        average_data.append(data[stats].mean())
        max_data.append(data[stats].max())
        min_data.append(data[stats].min())
        
    #Determines size of the graph
    plt.figure(figsize=(18,6))

    #Average Graph, sets the order, selects data source, sets labels for axes
    plt.subplot(1, 3, 1)
    plt.plot(average_data)
    plt.xlabel('Feature#')
    plt.ylabel('Average')

    #Max Graph, sets the order, selects data source, sets labels for axes
    plt.subplot(1, 3, 2)
    plt.plot(max_data)
    plt.xlabel('Feature#')
    plt.ylabel('Max')

    #Min Graph, sets the order, selects data source, sets labels for axes
    plt.subplot(1, 3, 3)
    plt.plot(min_data)
    plt.xlabel('Feature#')
    plt.ylabel('Min')

    #Prints Graph
    plt.show()
```

```python
#For Loop to run the created Functions for each dataset and print dataset names 
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
